### In the following table, we want to find, for each client, the date when their total savings is over 1000 €.


```python
ahorros_pd = \
    pd.DataFrame.from_dict(
        {'cliente': ['Cliente1', 'Cliente1', 'Cliente1', 'Cliente1', 'Cliente1', 'Cliente1',
                'Cliente2', 'Cliente2', 'Cliente2', 'Cliente2', 'Cliente2', 'Cliente2',
                'Cliente3', 'Cliente3', 'Cliente3', 'Cliente3', 'Cliente3', 'Cliente3',
                'Cliente4', 'Cliente4', 'Cliente4', 'Cliente4', 'Cliente4', 'Cliente4'],
        'fecha': [201901, 201902, 201903, 201904, 201905, 201906,
                201901, 201902, 201903, 201904, 201905, 201906,
                201901, 201902, 201903, 201904, 201905, 201906,
                201901, 201902, 201903, 201904, 201905, 201906],
        'ahorro': [150, 250, 0, 0, 0, 150,
                    250, 300, 320, 370, 250, 1200,
                    200, 100, 150, 0, 300, 300,
                    330, 290, 150, 180, 270, 350]})
```

```python
def calcular_suma(dic):

    # Estos son los valores iniciales que se irán modificando conforme avance el bucle.
    cliente_inicial = 'Cliente1'
    ahorro = 0
    logrado = 0

    res_cliente = []
    res_fecha = []
    res_ahorro = []

    for clave, valores in dic.items():
        filas = len(valores)
        # Aquí se obtiene el número de filas para iterar a continuación.

    for i in range(filas):
        if (cliente_inicial == dic['cliente'][i]) and (logrado == 0):       # Si el cliente de la iteración es el mismo que tenemos almacenado
            ahorro += dic['ahorro'][i]                                      # significa que podemos seguir sumando el ahorro.
            # print(ahorro, ahorros_pd['cliente'][i])
            if ahorro > 1000:                                               # Cuando el ahorro supera los 1000 €, guardamos los valores de cliente,
                print(dic['cliente'][i], dic['fecha'][i], ahorro)           # fecha y ahorro hasta ese momento.
                res_cliente.append(dic['cliente'][i])
                res_fecha.append(dic['fecha'][i])
                res_ahorro.append(ahorro)
                logrado = 1                                                 # Con la variable 'logrado' se evita que si el cliente es el mismo en la 
                # print(logrado)                                            # próxima iteración, no entremos al if inicial.

        elif cliente_inicial != dic['cliente'][i]:                          # Si el cliente de la iteración es diferente al almacenado, resetamos 
            cliente_inicial = dic['cliente'][i]                             # los valores de cliente inicial y logrado, para que en una próxima iteración
            logrado = 0                                                     # sea posible volver a entrar al if inicial. También se comienza el nuevo 
            ahorro = 0                                                      # cómputo de ahorro.
            ahorro += dic['ahorro'][i]
    return (res_cliente, res_fecha, res_ahorro)

calcular_suma(ahorros_pd)
```
